```python
model = load_model('/content/dogvscat_vgg16.h5')
```

```python
test_img = image.img_to_array(test_img)
test_img = test_img/255
test_img = np.expand_dims(test_img, axis = 0)
result = model.predict(test_img)
```

```python
a=np.argmax(result, axis=1)
```

```python
if a==0:
  print("Its a Cat!")
else:
  print("Its a Dog!")
```
